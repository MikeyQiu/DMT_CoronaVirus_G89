```python
import pandas as pd
df=pd.read_csv("data.csv")
```

```python
df1=df[df.AggregationType=='Annual mean / 1 calendar year']
df1
```

```python
df2=df1[df1.Validity=='Valid']
df2
```

```python
df3=df2[df2.Verification=='Verified']
df3=df3.loc[:,["CountryOrTerritory","ReportingYear","SamplingPoint_Latitude","SamplingPoint_Longitude","Pollutant","Unit","DataCoverage","DataCapture","AQValue"]]
df_SO2=df3[df3.Pollutant=='Sulphur dioxide (air)']
df_PM10=df3[df3.Pollutant=='Particulate matter < 10 µm (aerosol)']
df_CO=df3[df3.Pollutant=='Carbon monoxide (air)']
df_O3=df3[df3.Pollutant=='Ozone (air)']
df_NO2=df3[df3.Pollutant=='Nitrogen dioxide (air)']
df_PM25=df3[df3.Pollutant=='Particulate matter < 2.5 µm (aerosol)']
df_PM25
```

```python
result1 = pd.merge(df_SO2, df_PM10, how='left', on=['CountryOrTerritory','ReportingYear','SamplingPoint_Latitude', 'SamplingPoint_Longitude'])
result2 = pd.merge(df_CO, df_O3, how='left', on=['CountryOrTerritory','ReportingYear','SamplingPoint_Latitude', 'SamplingPoint_Longitude'])
result3 = pd.merge(df_NO2, df_PM25, how='left', on=['CountryOrTerritory','ReportingYear','SamplingPoint_Latitude', 'SamplingPoint_Longitude'])
# data=df_SO2.join(df_PM10,df_CO)
# #data = result3.join([result1, result2])
# data
print(len(result1),len(result2),len(result3))
result4=pd.merge(result2, result1, how='left', on=['CountryOrTerritory','ReportingYear','SamplingPoint_Latitude', 'SamplingPoint_Longitude'])
result5=pd.merge(result4, result3, how='left', on=['CountryOrTerritory','ReportingYear','SamplingPoint_Latitude', 'SamplingPoint_Longitude'])
result5.to_csv("combined.csv",index=False)
```

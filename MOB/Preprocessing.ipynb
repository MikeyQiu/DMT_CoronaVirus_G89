```python
import pandas as pd
Mobi=pd.read_csv("France_Mobi.csv", header=None)
Mobi=Mobi.T
Mobi.columns=['date','driving','transit','walking']
Corona=pd.read_csv("France_Corona.csv")
Mobi
```

```python
#Corona["date"][13]==Mobi["date"][4]
```

```python
df= pd.merge(Corona, Mobi, on=['date'])
df['days']=0
df
```

```python
num=1
# Here we only care for the date that has more than one hundreds cases
for i in range(len(df)):
    if df['total_cases'][i]>100:
        #print(df["date"][i])
        df["days"][i]=num
        num+=1
```

```python
df=df[df['days']>0]
df.to_csv('France.csv')
#df
```
